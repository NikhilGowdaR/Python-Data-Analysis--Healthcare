```python
# Step 3: Statistical analysis (Chi-square test)
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['thal'], df['target'])

# Perform Chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

print(f"Chi-square Test p-value: {p}")
```

```python
# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("There is a significant association between thalassemia and CVD.")
else:
    print("There is no significant association between thalassemia and CVD.")
```

```python
# Assuming you have loaded your data into a DataFrame called 'df'
# Replace 'column_name_for_target' with the actual column name for your target variable
X = df.drop(['target'], axis=1)
y = df['target']

# Step 1: Handle missing values (imputation)
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

# Example assuming 'Agegroup' is a column in X and is categorical
imputer = SimpleImputer(strategy='most_frequent')  # Use most frequent value for categorical variables
X_imputed = X.copy()
X_imputed[['Agegroup']] = imputer.fit_transform(X[['Agegroup']])  # Use double square brackets to keep it as a DataFrame

# Step 2: One-Hot Encode Categorical Variables
X_encoded = pd.get_dummies(X_imputed, columns=['Agegroup'])  # Replace 'Agegroup' with your actual categorical column

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Step 4: Scale the Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Example: Visualizing a Confusion Matrix for a Classification Model
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)

# Assuming you have a trained model 'model' and test data 'X_test_scaled', 'y_test'
y_pred = model.predict(X_test_scaled)
conf_matrix = confusion_matrix(y_test, y_pred)

sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Assuming 'data' is your dataset
# Assigning the 13 features to X and the last column to the classification predictor Y
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Splitting the data into Training & Test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Normalizing the data (standardization)
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

# Training Logistic Regression model on the training set
model1 = LogisticRegression(random_state=1)  # Get an instance of the model
model1.fit(x_train, y_train)  # Train/Fit the model

# Predicting on the test set
y_pred1 = model1.predict(x_test)

# Outputting classification report, which includes precision, recall, and F1-score
print(classification_report(y_test, y_pred1))
```
